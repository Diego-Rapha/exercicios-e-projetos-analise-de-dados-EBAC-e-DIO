```python
#Datas mais antigas da tiragem das cédulas/ou moedas, por Família (Espécie)
df.groupby('FAMÍLIA (ESPÉCIE)')['DATA'].min().sort_values(ascending=True)
```

```python
#Datas mais tardia da tiragem das cédulas/ou moedas de moedas, por Família (Espécie)
df.groupby('FAMÍLIA (ESPÉCIE)')['DATA'].max().sort_values(ascending=True)
```

```python
#Agrupamento de datas por Família (Espécie)(da menor data de cunhagem das moedas para a maior)
df.groupby('FAMÍLIA (ESPÉCIE)')['DATA'].min().sort_values(ascending=True)
```

```python
#LEVANTAMENTO DE CÉLULAS COM VALORES NULOS
df.isnull().sum()
```

```python
#AGRUPAMENTO DE MONTATE POR FAMÍLIA (ESPÉCIE)
df.groupby('FAMÍLIA (ESPÉCIE)')['MONTANTE'].sum()
```

```python
#Famílias (Espécies) de moedas e cédulas
df['FAMÍLIA (ESPÉCIE)'].unique()
```

```python
#Tiragem de moedas e cédulas por Família (Espécie)
df.groupby('FAMÍLIA (ESPÉCIE)')['TIRAGEM'].sum()
```

```python
#Tiragem das moedas por ano
df.groupby(df['DATA'].dt.year)['TIRAGEM'].sum()
```

```python
#Gráfico que mostra a evolução da tiragem de cédulas e moedas ao longo dos anos
df.groupby(df['DATA'].dt.year)['TIRAGEM'].sum().plot()
```

```python
#Agrupamento do montante(Valor das cédulas e moedas) produzido ao longo dos anos
df.groupby(df['DATA'].dt.year)['MONTANTE'].sum()
```

```python
#Gráfico que mostra a evolução do montante('Valor das cédulas e moedas') resultante da tiragem de cédulas e moedas ao longo dos anos.
#Pode-se observar a relação direta entre o gráfico anterior (Agrupamento de Tiragem por Ano) e o infracitado, dada a semelhança de evolução, apresentada nos mesmos.
df.groupby(df['DATA'].dt.year)['MONTANTE'].sum().plot()
```

OBS.: Nota-se que o gráfico de agrupamento do montante por ano, mostra uma evolução semelhante ao gráfico da tiragem por ano.

```python
#Um breve tratamento estatístico do valor montante('Valor das cédulas e moedas') resultante da tiragem de cédulas e moedas ao longo dos anos
df.groupby(df['DATA'].dt.year)['MONTANTE'].sum().describe()
```
